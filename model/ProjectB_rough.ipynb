```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("Downloads\\Data-Part1.csv") 
data = data.dropna()
data
```

```python
data_trunc = data.loc[:,'Vehicle Traffic':'Percent of Youth']
data_trunc_array=data_trunc.as_matrix()
```

```python
plt.plot(data_trunc_array[:,0],data_trunc_array[:,1], 'o')
plt.show
```

```python
data_trunc_TSNE = TSNE(n_components=2).fit_transform(data_trunc)
```

```python
data_trunc_TSNE
```

```python
plt.plot(data_trunc_TSNE[:,0],data_trunc_TSNE[:,1], 'o')
plt.show
```

```python
data_k = KMeans(n_clusters=15, random_state=0).fit(data_trunc_TSNE)
```

```python
data_k.labels_
```

```python
data_k.predict(array.reshape(-1,1)([10000, 3000, 35, 35000, 2, 8]))[0]
```

```python
data_labels = pd.DataFrame(data_k.labels_)
```

```python
data_total = pd.concat([data_labels, data], axis=1)
data_total
```

```python
is_val = data_total[0]==4
data_total_4 = data_total[is_val]
print(data_total_4['Intersection'].values)
```
